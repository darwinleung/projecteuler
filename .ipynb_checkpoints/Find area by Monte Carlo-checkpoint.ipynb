The function f below defines a region in two dimensions x and y, where each variable is
constrained to the [0,1] interval. It returns True if a point with coordinates (x, y) belongs to
the region, and False otherwise.
import math
def f(x, y):
return math.sqrt(x**2.0 + y**2.0) <= 1.0
For example, the point (0.1, 0.8) belongs to the region and therefore f returns True .
On the other hand, (0.9, 0.8) doesnâ€™t and the function returns False .

f(0.1, 0.8) True
f(0.9, 0.8) False

Let us assume that you don't know the mathematical definition of the region, i.e., the only
allowed operation is calling f to determine whether a given point (x, y) belongs to the region
or not. Write a Python program to calculate the area of the region.


```python
import math
def f(x, y):
    return math.sqrt(x**2.0 + y**2.0) <= 1.0
```

```python
import random
def findarea():
    n = 100000 # no. of iterations
    count = 0
    for i in range(n):
        if f(random.random(), random.random()):
            count += 1 # increment if point inside region
    return count*1.0/n # handling integer division
findarea()
```

```python
import numpy as np

n = 10000 # no. of iterations
random_pts_array = np.random.rand(n,2) # generate array of n pairs of random xy points
is_in_array = np.array([f(x,y) for x,y in random_pts_array]) # array of whether point is in region
print "Estimated area:",sum(is_in_array)*1.0/n # calculate proportion of points inside region
```

```python
#visualization
%matplotlib inline  
import matplotlib.pyplot as plt

in_xy = random_pts_array[is_in_array]
out_xy = random_pts_array[~is_in_array]

fig, ax = plt.subplots()
ax.scatter(in_xy[:,0],in_xy[:,1],c='g')
ax.scatter(out_xy[:,0],out_xy[:,1],c='r')
ax.set_aspect('equal')
fig.show()
```
